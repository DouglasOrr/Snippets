```python
import Box2D as B

world = B.b2World(gravity=(0, -10))
world.CreateStaticBody(
    position=[0, -10],
    shapes=B.b2PolygonShape(box=(50, 10)),
)
ball = world.CreateDynamicBody(position=(0, 4))
ball.CreatePolygonFixture(box=(1, 1), density=1, friction=.3)

world.ClearForces()
ball.ApplyForce([1, 0], ball.position, False)
for i in range(60):
    world.Step(1/60, 6, 2)
    print(tuple(ball.position))
```

```python
world = B.b2World(gravity=(0, -10))
ground_def = B.b2BodyDef()
ground_def.position.Set(0, -10)
ground_body = world.CreateBody(ground_def)
ground_shape = B.b2PolygonShape()
ground_shape.SetAsBox(50, 10)
ground_body.CreateFixture(shape=ground_shape)

ball_def = B.b2BodyDef()
ball_def.type = B.b2_dynamicBody
ball_def.position.Set(0, 4)
ball_body = world.CreateBody(ball_def)
ball_shape = B.b2PolygonShape()
ball_shape.SetAsBox(1, 1)
ball_fixture = B.b2FixtureDef()
ball_fixture.shape = ball_shape
ball_fixture.density = 1
ball_fixture.friction = 0.3
ball_body.CreateFixture(ball_fixture)

```
