```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_context('talk')
```

```python
loss = pd.read_json('results/q/training/loss.jsonl', lines=True)
outcome = pd.read_json('results/q/training/outcome.jsonl', lines=True)
plt.figure(figsize=(16,6))
loss.rolling(200, on='t').mean().dropna().plot(x='t', y='loss', style='b-', ax=plt.gca())
outcome[['t', 'duration']].rolling(10).mean().plot(x='t', y='duration', style='m-', ax=plt.gca().twinx())
```
